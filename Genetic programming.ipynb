<h2>Row formula</h2>

```python
random_data1 = []
#3000000
a1 = -1
a2 = 1
for i in range(4000): #(number of formula)
    l_difusivities_list = [1,2,3,4] #left diffusivities list
    r_diffusivities_list = [1,2,3,4] #right diffusivities list
    random.shuffle(l_difusivities_list)
    random.shuffle(r_diffusivities_list)
    ran = []
    a = random.randint(a1,a2)#coeficient 
    b = random.randint(a1,a2)#coeficient 
    c = random.randint(a1,a2)#coeficient 
    d = random.randint(a1,a2)#coeficient 
    e = random.randint(0,0)#math function
    f = random.randint(a1,a2)#coeficient 
    g = random.randint(a1,a2)#coeficient 
    h = random.randint(a1,a2)#coeficient 
    i = random.randint(a1,a2)#coeficient 
    j = random.randint(0,0) #math function
    s = random.randint(1,5)#random coefficient 
    t = random.randint(2,4)#random coefficient
    ran.append(a)
    ran.append(b)
    ran.append(c)
    ran.append(d)
    ran.append(e)
    ran.append(f)
    ran.append(g)
    ran.append(h)
    ran.append(i)
    ran.append(j)
    for i in range(4):
        ran.append(l_difusivities_list[i])
    for i in range(4):    
        ran.append(r_diffusivities_list[i])
    ran.append(s)
    ran.append(t)
    random_data1.append(ran)
```

<h3>The step of removing zero result in the formula </h3>

```python
random_data_0 = []
l = len(random_data1)
for i in range(l):
    c = 0
    c2= 0
    for j in range(4):
        if random_data1[i][j]==0:
            c = c + 1
        if random_data1[i][j+5]==0:
            c2 = c2 + 1 
    if c2<4 and c<4:
        random_data_0.append(random_data1[i])
        
```
